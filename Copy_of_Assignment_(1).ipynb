Replace all the tokens given in the (generic scheme)docx file according to a perticular company database by rendering data from a csv file.

```python
#Importing All necessary libraries
import pandas as pd
import numpy as np
import docx
from docx import Document
```

```python
#Stored all the tokens in an Array
tokens = ['<Company Name>','<ESOP Plan Name>','<Board Meeting Date>','<Shareholder Meeting Date>','<Number of ESOPs>','<Maximum ESOP Percentage>','<Cliff Period>','<Vesting Maximum Period>','<Resignation Vested Exercise Period>','<Resignation Unvested  Exercise Period>','<Termination vested  Exercise Period>','<Termination Unvested Exercise Period>','<Jurisdiction>']
```

```python
#mounted csv file
path = "/content/drive/MyDrive/ESOP_database - Sheet1 (1).csv"
```

```python


data = pd.read_csv(path)
data.head()
```

```python
#Converted all the columns into tokens and setting company name as index column
data.columns = tokens
data_idx = data.set_index('<Company Name>')
data_idx.head()
```

```python
#used iloc to select a perticular row or col from the dataset and selected comapny name to get its perticular data
Company_name = 'Boat'
idx = list(data_idx.index.values).index(Company_name)
comp_data = data.iloc[idx]
comp_data
```

```python
# created a function to repalce tokens with perticular data 
def replace_string(doc,old_text,new_text):
  paragraphs = list(doc.paragraphs)
  for t in doc.tables:
      for row in t.rows:
          for cell in row.cells:
              for paragraph in cell.paragraphs:
                  paragraphs.append(paragraph)
  for p in paragraphs:
      if old_text in p.text:
          inline = p.runs
          # Loop added to work with runs (strings with same style)
          for i in range(len(inline)):
              if old_text in inline[i].text:
                  text = inline[i].text.replace(old_text, new_text)
                  inline[i].text = text
                 
```

```python
filename = "/content/drive/MyDrive/Generic ESOP Scheme (1).docx"
doc = Document(filename)
file_new = (filename.split('.')[1])
for token in tokens:
  replace_string(doc,token,str(comp_data[token]))
doc.save('new.docx')


    
```
